# Recreating the GHC experiments using Graph_tools

## Experimenting with Graphs

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import itertools

import networkx as nx 
from networkx.algorithms.isomorphism import ISMAGS
```

```python
G = nx.Graph()

n = np.arange(10)
m = np.arange(10)

k = zip(n,m)
K_5 = nx.complete_graph(5)
K_4 = nx.complete_graph(4)
```

Computing some homomorphism spaces

```python
X = ISMAGS(K_5,K_4)

xit = X.subgraph_isomorphisms_iter()

xit_1 = itertools.islice(xit,1)

xpt = itertools.filterfalse(lambda x: x.get(0)!= 0, xit)

xpt_3 = itertools.islice(xpt,3)

list(xpt)

```
